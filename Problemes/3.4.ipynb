### 3.4 Resta Binària

Escriu una funció que resti dos nombres binaris que estan emmagatzemats en dues llises. Considera que es poden emmagatzemar en 4 bits i que si l'operació que volem és $x-y$, el minuend, $x$, ha de ser més gran que el subtrahehend, $y$.

```python
def resta_binaria(op1, op2):
    """
    Aquesta funció calcula la resta binaria donat dos nombres.
    
    parameters
    ----------
    op1: llista d'enters
    op2: llista d'enters
    
    Returns
    -------
    resultat: llista d'enters
    """
    decimal1 = 0
    decimal2 = 0
    emportar = 0
    resultat = []
    result = []
    for i in range(4):
      decimal1 += op1[i]*(2**(3-i))
      decimal2 += op2[i]*(2**(3-i))
    if decimal2 > decimal1:
      print("El primer numero ha de ser major al segon.")
      return -1
    for i in range(4):
      if emportar == 1:
        decimal2[3-i] += 1
        emportar = 0
      operacio = op1[3-i] - op2[3-i]
      match operacio:
        case 0 | 1:
          resultat.append(operacio)
        case -1:
          emportar = 1
          resultat.append(0)
    for i in range(4):
      result.append(resultat[3-i])
    return result
 
```

```python
assert resta_binaria([0,1,0,1], [0,0,0,1]) == [0,1,0,0]
```
