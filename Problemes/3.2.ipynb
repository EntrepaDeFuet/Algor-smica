### 3.2 El Màxim Comú Divisor i les seves aplicacions

L'algorisme d'Euclides per trobar el màxim comú divisor de dos nombres $a$ i $b$ és el següent:
```python
def gcd(a,b):
    while a:
        a, b = b%a, a
    return b
```

Fent servir aquest algorisme, escriu una funció que simplifiqui una fracció a la seva expressió irreductible.

````python
def gcd(a,b):
    while a:
        a, b = b%a, a
    return b
 
def reduir_fraccio(numerador, denominador):
    """
    Funció que retorna l'expressio irreductible d'una fracció
    
    Parameters
    ----------
    numerador: int
    denominador: int
    
    Returns
    -------
    numReduit: int
    denReduit: int
    """
    mcd = gcd(numerador,denominador)
    numReduit = numerador / mcd
    denReduit = denominador / mcd
    return (numReduit,denReduit)
````

````python
assert reduir_fraccio(12, 8) == (3,2)
````
