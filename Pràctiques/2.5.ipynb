### 2.5 Capicua

Escriu una funció que donada una cadena de caràcters determini si és capicua (es llegeixen igual en qualsevol direcció).

```python
def es_capicua(paraula):
    """
    Aquesta funció determina si una paraula és capicua.
    
    Parameters
    ----------
    paraula: string
        La paraula a determinar
    
    Returns
    -------
    b: bool
        true si la paraula és capicua, false altrament
    """ 
    b = True
    for i in range(int(len(paraula)/2)):
        if paraula[i] != paraula[len(paraula)-i-1]:
            b = False
    
    return b
```

```python
assert es_capicua("llull") == True
assert es_capicua("llullu") == False
```

Escriu una funció que, donada una cadena de caràcters, determini si en aplicar rotacions de lletres, és capicua. Per resoldre aquest problema has d'anar creant les paraules rotades progressivament i validant si són capicua amb l'algorisme anterior.

Una rotació consisteix a situar l'última lletra en la primera posició o viceversa. Per exemple, en la cadena 'ABCD' les rotacions serien: ['DABC', 'CDAB', 'BCDA', 'ABCD'].

```python
def es_capicua_rotacions(paraula): 
    """
    Aquesta funció determina si una paraula és capicua aplicant rotacions.
    
    Parameters
    ----------
    paraula: string
        La paraula a determinar
    
    Returns
    -------
    b: bool
        true si la paraula és capicua, false altrament
    """
    contador = 0
    control =  False
    b = es_capicua(paraula)
    if b != True:
        while control == False:
             paraula = rotar(paraula)
             contador += 1
             b = es_capicua(paraula)
             control = b
             if contador == len(paraula):
                 control = True
    
    
    return b
```

```python
assert es_capicua_rotacions("BAABCC") == True
```

```python
def rotar(paraula):
    lletra = paraula[0]
    paraula = paraula[1:]+lletra
    return paraula
```
