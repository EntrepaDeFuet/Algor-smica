```python
import math

def setmanes_segons():
    """
    Aquesta funció troba tots els nombres de setmanes N que tenen 
    un nombre de segons S tal que S = K! per algun k < 100
    
    Returns
    -------
    res: list
    """
    n_setmanes = []
    setmana = 604800
    for i in range(100):
        if math.factorial(i)%setmana == 0:
            n_setmanes.append(int(math.factorial(i)/setmana))
            print(n_setmanes)
  
    return n_setmanes
```

```python
assert setmanes_segons() == [6, 66, 792]
```

```python
import math 

def llistafactorial (n):
    llista =[]
    num = 1
    while math.factorial(num) < n:
       llista.append(math.factorial(num)+1)
       num += 1
    return llista
```
