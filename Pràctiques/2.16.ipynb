### 2.16  Nombres ambiciosos

+ 6 és un nombre perfecte.

```python
def ambicios_recursiu(num):
    """
    Aquesta funció comprova si un nombre es ambiciós.
    Implementació recursiva
    
    Parameters
    ----------
    num: int
        El número que es vol comprovar
    
    Returns
    -------
    b: bool
        Si el nombre d'entrada és un nombre ambiciós
    
    """
    it = num*1
    """Variable iterativa. Es fa la multipicació amb l'objectiu de no modificar el num introduït"""
    b = False
    """Variable de Control."""
    while it > 1:
        if b == True:
            it = 1
        else:
            if ambicios(it)==True:
                b = True
            else:
                it = sumaDivisors(it)
    return b
```

```python
assert ambicios_recursiu(25) == True  
assert ambicios_recursiu(100) == False 
assert ambicios_recursiu(143) == True 
```

```python
def divisorsPropis (num):
    """Aquesta funció busca els divisors propis d'un número"""
    divisors = []
    for i in range(1,num):
        residu = num % i
        if residu == 0:
            divisors.append(i)
    return divisors
```

```python
def esPerfecte (num):
    suma = sumaDivisors(num)
    if suma == num:
        return True
    else:
        return False
```

```python
def sumaDivisors(num):
    divisors = divisorsPropis(num)
    suma = 0
    for i in divisors:
        suma += i
    return suma
```
