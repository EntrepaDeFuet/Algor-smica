### 4.1 Acronims

Escriu una funció que a partir d'una frase sense punts, comes, accents, números ni cap signe de puntuació, imprimeixi l'acrònim corresponent.

```python
def acronim(frase):
    """
    Aquesta funció retorna l'acrònim d'una frase.
    
    Parameters
    ----------
    frase: string
    
    Returns
    -------
    acronim: string
    """
    acronim = ''
    llista = frase.split(' ')
    for i in llista:
        primera = True
        for lletra in i:
            
            if primera == True:
                acronim += lletra.upper()
                primera = False
    return acronim
```

```python
assert acronim("Hola que tal") == 'HQT'
```

```python
frase = input('Digues la frase:')

print(acronim(frase))
```
